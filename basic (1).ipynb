```python
categorical_columns = ['mainroad','basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: 1 if x == 'yes' else 0)
```
